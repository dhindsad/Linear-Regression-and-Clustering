```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

data= pd.read_csv('/Users/DeepDhindsa/Desktop/COSC4426/penguins.csv')
data.head()
```

```python
kmean_data=data.copy()
kmean_data=kmean_data.fillna(0)
X_kmean= kmean_data[["bill_length_mm", "bill_depth_mm"]]
kmeans= KMeans(n_clusters=3, init='k-means++', random_state =11)
kmeans.fit(X_kmean)
centers=kmeans.cluster_centers_
sns.scatterplot(x = data["bill_depth_mm"], y=data["bill_depth_mm"], hue=data["species"])
sns.scatterplot(x = centers[:,0], y=centers[:,1], color="DarkRed", s=100, alpha=0.85)
```
